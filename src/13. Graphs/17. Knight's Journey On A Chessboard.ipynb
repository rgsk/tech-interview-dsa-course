````python
from typing import List
from collections import deque

class Cell:
    def __init__(self, x: int, y: int):
        self.x = x
        self.y = y

class Solution:
    def minMovesRequired(self, n: int, start: Cell, end: Cell) -> int:
        ops = [(2, 1), (2, -1), (-2, 1), (-2, -1),
               (1, 2), (-1, 2), (1, -2), (-1, -2)]
        
        start_x, start_y = start.x - 1, start.y - 1
        end_x, end_y = end.x - 1, end.y - 1
        
        visited = [[False] * n for _ in range(n)]
        q = deque()
        q.append((start_x, start_y))
        visited[start_x][start_y] = True
        
        moves = 0
        
        while q:
            for _ in range(len(q)):
                r, c = q.popleft()
                
                if r == end_x and c == end_y:
                    return moves
                
                for dx, dy in ops:
                    nr, nc = r + dx, c + dy
                    if 0 <= nr < n and 0 <= nc < n and not visited[nr][nc]:
                        visited[nr][nc] = True
                        q.append((nr, nc))
            
            moves += 1
        
        return -1  # Unreachable
````

    q.push({start.x, start.y});
    
    vector<vector<bool>> visited(n, vector<bool>(n, false));
    visited[start.x][start.y] = true;
    
    int moves = 0;
    
    while (!q.empty()) {
        int size = q.size();
        for (int i = 0; i < size; i++) {
            auto [r, c] = q.front();
            q.pop();
            
            if (r == end.x && c == end.y) {
                return moves;
            }
            
            for (auto op : ops) {
                int nr = r + op.first;
                int nc = c + op.second;
                
                if (nr >= 0 && nc >= 0 && nr < n && nc < n && !visited[nr][nc]) {
                    q.push({nr, nc});
                    visited[nr][nc] = true;
                }
            }
        }
        moves++;
    }
    
    return -1;
}
