```python
class Queue:
	def __init__(self, capacity=0):
		self.arr = [0 for _ in range(capacity)]
		self.l = 0
		self.r = 0

	def isEmpty(self) -> bool:
		return self.size() == 0

	def size(self) -> int:
		return self.r - self.l

	def front(self) -> int:
		if self.isEmpty(): return -1
		return self.arr[self.l]

	def back(self) -> int:
		if self.isEmpty(): return -1
		return self.arr[self.r - 1]

	def push(self, element: int) -> None:
		self.arr[self.r] = element
		self.r += 1

	def pop(self) -> None:
		self.l += 1
```

```python
class Stack:
	def __init__(self, capacity=0):
		self.queue = Queue(capacity)

	def isEmpty(self) -> bool:
		return self.size() == 0

	def size(self) -> int:
		return self.queue.size()

	def top(self) -> int:
		return self.queue.front()

	def push(self, element: int) -> None:
		self.queue.push(element)
		for i in range(self.queue.size() - 1):
			self.queue.push(self.queue.front())
			self.queue.pop()
			

	def pop(self) -> None:
		self.queue.pop()
```

```python
def main():
    stack = Stack(10)

    print(stack.top(), stack.isEmpty(), stack.size())

    stack.push(5)
    print(stack.top(), stack.isEmpty(), stack.size())

    stack.push(6)
    print(stack.top(), stack.isEmpty(), stack.size())

    stack.push(7)
    print(stack.top(), stack.isEmpty(), stack.size())

    stack.pop()
    print(stack.top(), stack.isEmpty(), stack.size())

    stack.pop()
    print(stack.top(), stack.isEmpty(), stack.size())

    stack.pop()
    print(stack.top(), stack.isEmpty(), stack.size())

main()
```
